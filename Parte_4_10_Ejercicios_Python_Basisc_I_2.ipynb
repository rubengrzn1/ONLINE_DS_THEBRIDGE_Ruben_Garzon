## Ejercicio 1

1. Crea dos variables numericas: un `int` y un `float`
2. Comprueba sus tipos
3. Sumalas en otra nueva
4. ¿De qué tipo es la nueva variable?
5. Elimina las dos primeras variables creadas

```python
#Creación de 2 variables y comprobación del tipo de cada una de ellas.
numero_1 = 2
numero_2 = 3.56
print(type(numero_1))
print(type(numero_2))

#Suma de las 2 variables y ademas comprobación del tipo de suma.
suma = numero_1 + numero_2
print(suma)
print(type(suma))

#Eliminar las 2 primeras variables.
del numero_1, numero_2

#Comprobación de que la variable ya no existe:
print(numero_1)
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
base = input("Centimetros de base ")
altura = input("Centimetros de altura ")
base_1= float(base.replace(",", "."))
altura_1= float(altura.replace(",", "."))
paralelogramo = base_1 * altura_1
print(paralelogramo)
```
