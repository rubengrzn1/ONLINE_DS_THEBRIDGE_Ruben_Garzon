```python
# Declarar y asignar una variable entera
entero = 4
# Declarar y asinar una variale real o float
real = 14.3
# Sumar las variables y mostrar el resultado
resultado = entero + real
print(resultado)
# Mostrar que tipo es el resultado
print(type(resultado))
# Multiplicar ambas variables y mostrar el resultado
multi = entero * real
print(multi)
print(type(multi))
# Elevar una a la otra
potencia = entero ** real
print(potencia)
print(type(potencia))
# Dividir una por otra y mostrar ademas el resto y el cociente
divi = entero / real
resto = entero // real
cociente = entero % real
print(divi,"\n",resto,"\n",cociente)
```
