## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1
if A==B:
    print("True")
else:
    print("False")

if A!=C:
    print("True")
else:
    print("False")
if A>C:
    print("True")
else:
    print("False")
if C<=A:
    print("True")
else:
    print("False")
if B!=C:
    print("True")
else:
    print("False")
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

3. Dentro del paréntesis hay 3 False y un True, por lo tanto el resultado es True, pero el tema es que como fuera hay un y False, el resultado final sería **False.**

4. Este caso sería similar al caso 1, donde not false, se convierte en True y el not True, se convierte en False. Por lo tanto, la secuencia sería True and True and False, conclusión, **False**.
