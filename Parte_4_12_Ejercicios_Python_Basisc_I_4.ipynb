## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

1. Calcula el máximo de la lista: [4, 6, 8, -1]
2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
#Sacar el valor máximo de la lista
string = [4,6,8,-1]
print(max(string))
#Sumar elementos de la lista anterior
print(sum(string))
#Poner el siguiente número con 3 decimales
num_decimales = 3
print(round(63.451256965, num_decimales))
#Valor absoluto
print(abs(-74))
```

## Ejercicio 2
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

```python
frase = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
print(frase.upper())
print(frase.lower())
print(frase.title())
print(frase.split())
print(frase.replace(",",";"))
print(frase.replace("a", ""))
```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
#Creación de lista
lista = [1,2,3]
#Añadido un cuarto elemento
lista.append(22)
print(lista)
#calcular la suma de todo
suma_total= sum(lista)
print(suma_total)
#Eliminar el segundo elemento de la lista
lista.pop(1)
print(lista)
#Añade un elemento en la posición 3 de la lista
lista.insert(3, 100)
print(lista)
#Otra lista y concatenar con la anterior
lista2 = [1,2,3,4]
lista_final = lista + lista2
print(lista_final)
#Ordenar la lista de menor a mayor
lista_final.sort()
print(lista_final)
#Ordenar la lista de mayor a menor
lista.sort(reverse=True)
print(lista)

```
